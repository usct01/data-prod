```python
import matplotlib.pyplot as plt

# Define working hours (example)
india_start = 11.5  # IST working hours start (considering 30 min buffer)
india_end = 22.5  # IST working hours end (considering 30 min buffer)
us_start = 8  # EST working hours start
us_end = 17  # EST working hours end 

# Create a time axis
time_axis = range(0, 24)

# Define time zones (shaded areas)
plt.fill_between(time_axis, india_start, india_end, alpha=0.2, color='green', label='India (IST)')
plt.fill_between(time_axis, us_start, us_end, alpha=0.2, color='blue', label='US (EST)')

# Label the axes
plt.xlabel('Time (Hours)')
plt.ylabel('Timezone')

# Add a title and legend
plt.title('Working Hour Overlap (IST & EST)')
plt.legend()

# Display the chart
plt.grid(True)
plt.show()
```
